```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader


# Define a custom dataset class
class MolecularDataset(Dataset):
    def __init__(self, selfies_data, labels):
        self.selfies_data = selfies_data
        self.labels = labels

    def __len__(self):
        return len(self.selfies_data)

    def __getitem__(self, idx):
        selfies = self.selfies_data[idx]
        label = self.labels[idx]
        return selfies, label

# Define the LSTM-based regression model
class MolecularLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=1):
        super(MolecularLSTM, self).__init__()
        self.embedding = nn.Embedding(input_size, hidden_size)
        self.lstm = nn.LSTM(hidden_size, hidden_size, num_layers=num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = self.embedding(x)
        lstm_out, _ = self.lstm(x)
        output = self.fc(lstm_out[:, -1, :])
        return output

batch_size = 32
train_dataset = MolecularDataset(X_train.reset_index(drop=True), y_train.reset_index(drop=True))
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = MolecularDataset(X_test.reset_index(drop=True), y_test.reset_index(drop=True))
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)

optimization_dataset = MolecularDataset(X_optimization.reset_index(drop=True), y_optimization.reset_index(drop=True))
optimization_loader = DataLoader(optimization_dataset, batch_size=batch_size, shuffle=True)

val_dataset = MolecularDataset(X_val.reset_index(drop=True), y_val.reset_index(drop=True))
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=True)
```

```python

def selfies_to_tensor(selfies_data):
    # Convert selfies to tokens
    tokenized_selfies = [list(sf.split_selfies(s)) for s in selfies_data]

    # Find the length of the longest sequence
    max_len = max(len(seq) for seq in tokenized_selfies)

    # Create a vocabulary mapping for the tokens
    alphabet = list(sf.get_alphabet_from_selfies(selfies_data))
    vocabulary =  {**{'[nop]': 0}, **{alphabet[i]:i+1 for i in range(len(alphabet)) }}

    # Convert tokens to numerical indices
    numerical_selfies = [
        [vocabulary[token] for token in seq] + [vocabulary['[nop]']] * (max_len - len(seq))
        for seq in tokenized_selfies
    ]

    # Convert numerical selfies to a tensor
    tensor = torch.tensor(numerical_selfies, dtype=torch.long)

    return tensor

# Example usage
selfies_data1 = ['[C][C][O]', '[C][O][H]']
tensor = selfies_to_tensor(selfies_data1)

print("Original Selfies:")
print(selfies_data1)
print("\nPadded and Numerical Selfies Tensor:")
print(tensor)
```

```python
class MolecularLSTM2(nn.Module): #need to change num_kayers to be dynamic and the layers to be created dynamically
    def __init__(self, input_size, hidden_size, num_layers=1):
        super(MolecularLSTM2, self).__init__()
        self.embedding = nn.Embedding(input_size, hidden_size)

        # Additional LSTM layers
        self.lstm1 = nn.LSTM(hidden_size, hidden_size, num_layers=1, batch_first=True)
        self.lstm2 = nn.LSTM(hidden_size, hidden_size, num_layers=1, batch_first=True)

        # Linear layer
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = self.embedding(x)

        # Apply the first LSTM layer
        lstm_out1, _ = self.lstm1(x)

        # Apply the second LSTM layer
        lstm_out2, _ = self.lstm2(lstm_out1)

        # Final linear layer
        output = self.fc(lstm_out2[:, -1, :])
        return output
```

```python
# Hyperparameters
alphabet = list(sf.get_alphabet_from_selfies(selfies_data))
vocabulary =  {**{'[nop]': 0}, **{alphabet[i]:i+1 for i in range(len(alphabet)) }}
#input_size = len(sf.get_alphabet_from_selfies(selfies_data)) + 1# for nop  # Replace with the actual size of your vocabulary
input_size = len(vocabulary)
hidden_size = int(input_size/2)
num_epochs = 1
learning_rate = 0.001
num_epochs = 10
batch_size = 32
num_layers = 1

print(input_size, hidden_size)

model = MolecularLSTM2(input_size, hidden_size, num_layers)


def train_model(model, train_loader, num_epochs, learning_rate=0.001, gradient_clip=None):

    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    if gradient_clip:
        # Gradient clipping to prevent exploding gradients
        clip_value = gradient_clip
        for p in model.parameters():
            p.grad.data.clamp_(-clip_value, clip_value)

    # Training loop
    for epoch in range(num_epochs):
        for selfies, labels in train_loader:
            optimizer.zero_grad()

            # Convert selfies to tensor (replace with your preprocessing logic)
            # For example, you can use a library like RDKit to convert selfies to molecular graphs
            # Here, we assume you have a function selfies_to_tensor
            selfies_tensor = selfies_to_tensor(selfies)

            # Forward pass
            outputs = model(selfies_tensor)

            # Calculate loss
            loss = criterion(outputs, labels.view(-1, 1).float())
            # Backward pass and optimization
            loss.backward()
            # Gradient clipping to prevent exploding gradients

            optimizer.step()

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')

    return model

trained_model  = train_model(model, train_loader, num_epochs, learning_rate=0.001)
# Save the trained model
#torch.save(model.state_dict(), 'molecular_regression_model.pth')
```

```python
from sklearn.metrics import r2_score

def eval(model, test_loader):

    criterion = nn.MSELoss()
    model.eval()  # Set the model to evaluation mode

    total_loss = 0.0
    total_predictions = []
    total_labels = []

    with torch.no_grad():  # Disable gradient computation during evaluation
        for selfies, labels in test_loader:
            # Convert selfies to tensor
            selfies_tensor = selfies_to_tensor(selfies)

            # Forward pass
            outputs = model(selfies_tensor)

            # Calculate loss
            loss = criterion(outputs, labels.view(-1, 1).float())

            total_loss += loss.item()

            # Convert tensors to numpy arrays for R^2 score calculation
            predictions_np = outputs.detach().numpy().flatten()
            labels_np = labels.numpy().flatten()

            total_predictions.extend(predictions_np)
            total_labels.extend(labels_np)

    average_loss = total_loss / len(test_loader)
    print(f'Average Loss on Test Set: {average_loss}')

    # Calculate R^2 score
    r2 = r2_score(total_labels, total_predictions)
    print(f'R^2 Score on Test Set: {r2}')
    return r2, predictions_np
```
